```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a dictionary to store the precision values
data = {
    'Model': ['FNN'] * 10 + ['CNN'] * 10 + ['RNN'] * 10,
    'Category': list(range(10)) * 3,
    'Precision': [
        0.82, 0.84, 0.81, 0.84, 0.77, 0.90, 0.82, 0.91, 0.89, 0.89,  # FNN
        0.82, 0.83, 0.79, 0.84, 0.77, 0.87, 0.78, 0.94, 0.86, 0.85,  # CNN
        0.85, 0.85, 0.78, 0.88, 0.84, 0.63, 0.89, 0.95, 0.93, 0.90   # RNN
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the style and figure size
sns.set_style("whitegrid")
plt.figure(figsize=(16, 8))

# Create the bar plot
ax = sns.barplot(x='Category', y='Precision', hue='Model', data=df)

# Customize the plot
plt.title('Precision Comparison of Neural Network Models by Category', fontsize=18)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Precision', fontsize=14)

# Move legend outside the plot
plt.legend(title='Model', title_fontsize='14', fontsize='12', bbox_to_anchor=(1, 1), loc='upper left')

# Adjust layout to prevent cutting off the legend
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a dictionary to store the recall values
data = {
    'Model': ['FNN'] * 10 + ['CNN'] * 10 + ['RNN'] * 10,
    'Category': list(range(10)) * 3,
    'Recall': [
        0.87, 0.84, 0.82, 0.84, 0.78, 0.82, 0.84, 0.90, 0.88, 0.87,  # FNN
        0.85, 0.83, 0.81, 0.83, 0.75, 0.85, 0.84, 0.87, 0.88, 0.86,  # CNN
        0.85, 0.86, 0.87, 0.82, 0.65, 0.93, 0.80, 0.90, 0.84, 0.88   # RNN
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the style and figure size
sns.set_style("whitegrid")
plt.figure(figsize=(16, 8))

# Create the bar plot
ax = sns.barplot(x='Category', y='Recall', hue='Model', data=df)

# Customize the plot
plt.title('Recall Comparison of Neural Network Models by Category', fontsize=18)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Recall', fontsize=14)

# Move legend outside the plot
plt.legend(title='Model', title_fontsize='14', fontsize='12', bbox_to_anchor=(1, 1), loc='upper left')

# Adjust layout to prevent cutting off the legend
plt.tight_layout()

# Show the plot
plt.show()
```
